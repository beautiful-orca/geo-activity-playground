```python
from PIL import Image, ImageDraw
import numpy as np
import pandas as pd
```

```python
ts = pd.read_parquet(
    "~/Dokumente/Karten/Playground-Strava-API/Cache/Activity Timeseries/10159108471.parquet"
)
ts
```

```python
time_diff = (ts["time"] - ts["time"].shift(1)).dt.total_seconds()
```

```python
sorted(dict(zip(*np.unique(time_diff, return_counts=True))).items())
```

```python
jump_indices = time_diff >= 15
segments = np.cumsum(jump_indices)
segments
```

```python
time = pd.Series([1, 2, 3, 4, 60, 61, 62])
time_diff = time - time.shift(1)
jump_indices = time_diff >= 15
pd.DataFrame({
    "time": time,
    "segment": np.cumsum(jump_indices)
})
```
